Load and Explore the Dataset

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/Financial Analytics data.csv')

# Display the first few rows of the dataset
data.head(488)
```

```python
data.shape
```

Prepare the Data

```python
# Remove the unnecessary column
data = data.drop(columns=['Unnamed: 4'])

# Check for missing values
missing_values = data.isnull().sum()

# Ensure correct data types
data['Mar Cap - Crore'] = pd.to_numeric(data['Mar Cap - Crore'], errors='coerce')
data['Sales Qtr - Crore'] = pd.to_numeric(data['Sales Qtr - Crore'], errors='coerce')

# Check data types
data_types = data.dtypes

missing_values, data_types
```

Handle Missing Values

```python
# Remove rows with missing values
cleaned_data = data.dropna()

# Check the shape of the cleaned dataset
cleaned_data_shape = cleaned_data.shape

cleaned_data_shape
```

The cleaned dataset now has 365 rows and 4 columns. This data is ready to be analyzed in Tableau.

```python
cleaned_data.to_csv('cleaned_financial_data.csv', index=False)
```
